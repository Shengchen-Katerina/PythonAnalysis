### Linear Regression: train and test datasets

```python
#Split train and test datasets
from sklearn.model_selection import train_test_split
```

```python
data1.columns
```

```python
#Linear relationship
#1. The dependent variable is [price]: 
#      - positive relationships with [bathrooms],[sqft_living],[sqft_above],and [sqft_living15]
#2. All categorical fields except floors have positive relationship with dependent variable [price].
#3. No obvious relationship between built year and price
#4. Renovated property has a higher price


#Multicollinearity -- 
#1. sqft_above and sqft_living has a strong positive relationship (0.88)
#2. sqft_living amd sqft_living15 has a strong positive relationship (0.76)
#3. sqft_above amd sqft_living15 has a strong positive relationship (0.73)
#4. sqft_lot amd sqft_lot15 has a strong positive relationship (0.72)

X = data1.drop(['price','date','zipcode','Month_year','yr_renovated','yr_built',
                'bedrooms','floors','sqft_above','sqft_living15','sqft_lot15'],axis=1)
y = data1['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=42)
X_train
```

```python
#Train the algorithm
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
lr1 = reg.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(reg.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

```python
reg.intercept_
```

### Evaluate model performance

```python
import statsmodels.api as sm
X_train_lm = sm.add_constant(X_train)

lr_1 = sm.OLS(y_train, X_train_lm).fit()

lr_1.summary()
```

```python
#Predict for Unknown values

y_predicted = reg.predict(X_test)
residuals= y_test - y_predicted
```

```python
#Evaluate model by charts
fig, axs = plt.subplots(2,2, figsize=(10,10))

plt1 = axs[0][0].scatter(x=y_test, y=y_predicted)
axs[0][0].set_title('Fitted vs Predicted')
axs[0][0].set_xlabel('Price - test')
axs[0][0].set_ylabel('Price - predicted')

#detect heteroscedasticity
plt2 = axs[0][1].scatter(x=y_test, y=residuals)
axs[0][1].hlines(0, xmin=-3, xmax=3)
axs[0][1].set_title('Fitted Values vs Residuals')
axs[0][1].set_xlabel('Price - test (fitted)')
axs[0][1].set_ylabel('Residuals')
#The residuals don't have constant variance at every level of x:transform the dependent variable

#Normality
axs[1][0].scatter(x=sorted(np.random.randn(len(residuals))), 
                  y=sorted(residuals))
axs[1][0].set_title('QQ plot of Residuals')
axs[1][0].set_xlabel('Std. normal z scores')
axs[1][0].set_ylabel('Residuals')
#The residuals of the model aren't normally distributed

sns.distplot(residuals, ax=axs[1,1])
axs[1][1].set_title('Histogram of residuals')
plt.tight_layout()
```

```python
#Predict for Unknown values

y_predicted3 = reg3.predict(X_test3)
residuals3= y_test3 - y_predicted3
```

```python
#Evaluate model by charts
fig, axs = plt.subplots(2,2, figsize=(10,10))

plt1 = axs[0][0].scatter(x=y_test3, y=y_predicted3)
axs[0][0].set_title('Fitted vs Predicted')
axs[0][0].set_xlabel('Price - test')
axs[0][0].set_ylabel('Price - predicted')
#Fitted vs predicted chart shows a positive correlation between the predictions and actual values

#detect heteroscedasticity
plt2 = axs[0][1].scatter(x=y_test3, y=residuals3)
axs[0][1].hlines(0, xmin=-3, xmax=3)
axs[0][1].set_title('Fitted Values vs Residuals')
axs[0][1].set_xlabel('Price - test (fitted)')
axs[0][1].set_ylabel('Residuals')
#The residuals have constant variance at every level of x

#Normality
axs[1][0].scatter(x=sorted(np.random.randn(len(residuals3))), 
                  y=sorted(residuals3))
axs[1][0].set_title('QQ plot of Residuals')
axs[1][0].set_xlabel('Std. normal z scores')
axs[1][0].set_ylabel('Residuals')
#The residuals of the model are normally distributed

sns.distplot(residuals3, ax=axs[1,1])
axs[1][1].set_title('Histogram of residuals')
plt.tight_layout()
```
